```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# read the dataset
dataset = pd.read_csv("/content/Melbourne_housing_FULL.csv") 
dataset.head()
```

```python
# printing unique values
dataset.nunique()
```

```python
dataset.shape
```

```python
# Columns to use
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom',
               'Car','Landsize','BuildingArea','Price']

# Build the dataset
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
# Check for na values
dataset.isna().sum()
```

```python
# Handling Missing Values
# Replacing NA values with Zero
cols_fill_to_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_fill_to_zero] = dataset[cols_fill_to_zero].fillna(0)
dataset.isna().sum()
```

```python
# Fill NA with mean
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.head()
```

```python
dataset.isna().sum()
```

```python
# Perform one hot encoding
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset.Price
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=2)
```

```python
# Perform linear_regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test, y_test)
```

#### Using Lasso Regression

```python
from sklearn import linear_model 

lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```
