```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Plot (x,y) on that axes and set the labels and titles

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
# Plot the figure

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')

fig # Show figure
```

```python
#Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
axes1.plot(x,z)
axes1.set_xlabel('x')
axes1.set_ylabel('z')

axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom')
axes2.set_xlim(20,22)
axes2.set_ylim(30,50)

fig # Show figure
```

```python
#Use plt.subplots(nrows=1, ncols=2) to create the plot below.
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#plot (x,y) and (x,z) on the axes
axes[0].plot(x,y,color="b",lw=3,ls="--")
axes[1].plot(x,z,color="r",lw=3,ls="-")
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="b",lw=5,ls="-")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x,z,color="r",lw=5,ls="--")
axes[1].set_xlabel("x")
axes[1].set_ylabel("z")
fig # plot the graph
```
