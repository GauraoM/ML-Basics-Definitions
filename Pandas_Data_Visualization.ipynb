```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df3 = pd.read_csv('df3')
```

```python
df3.info()
```

```python
df3.head()
```

```python
# Scatter plot
df3.plot.scatter(x='a',y='b',c='r',s=50,figsize=(12,3))
```

```python
#Create a histogram of the 'a' column
df3['a'].plot.hist()
```

```python
#Using style sheets to set the style to 'ggplot' and redo the histogram above
plt.style.use('ggplot')
df3['a'].plot.hist(bins=25,alpha=0.5)
```

```python
# Creating a boxplot comparing the a and b columns
df3[['a','b']].plot.box(color='r')
```

```python
#Creating a kde plot of the 'd' column
df3['d'].plot.density(color='r')
```

```python
df3['d'].plot.density(color='r',lw=5, ls='--')
```

```python
#Creating an area plot of all the columns for just the rows up to 30
df3.loc[0:30].plot.area(alpha=0.4)
```

```python
#display the legend outside of the plot 
f = plt.figure()
df3.loc[:30].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=[1.0,0.5])
plt.show()
```
