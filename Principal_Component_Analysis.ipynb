```python
# Replacing object with integer
df.ExerciseAngina.replace({'N':0,'Y':1}, inplace=True)
df
```

```python
# Replacing object with integer
df.RestingECG.replace({'Normal':1, 'ST':2, 'LVH':3}, inplace=True)
df.ST_Slope.replace({'Down':1, 'Flat':2, 'Up':3}, inplace=True)
df.head()
```

```python
# performing label encoding
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
X = df.drop('HeartDisease',axis=1)
y = df.HeartDisease
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler

scaled = StandardScaler()
X_scaled = scaled.fit_transform(X)
X_scaled
```

```python
# Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# create an object and perform fitting and check accuracy
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test, y_test)
```

```python
# create an object and perform fitting and check accuracy
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test, y_test)
```

```python
# create an object and perform fitting and check accuracy
svm = SVC()
svm.fit(X_train,y_train)
svm.score(X_test, y_test)
```

#### PCA

```python
# import model and perform pca
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
```

```python
# Split the data
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2)
```

```python
# create an object and perform fitting and check accuracy
lr = LogisticRegression()
lr.fit(X_train_pca,y_train)
lr.score(X_test_pca, y_test)
```

```python
# create an object and perform fitting and check accuracy
rf = RandomForestClassifier()
rf.fit(X_train_pca,y_train)
rf.score(X_test_pca, y_test)
```

```python
# create an object and perform fitting and check accuracy
svm = SVC()
svm.fit(X_train_pca,y_train)
svm.score(X_test_pca, y_test)
```
